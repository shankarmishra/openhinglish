# OpenHinglish — 60-second quickstart

[OpenHinglish](https://github.com/shankarmishra/openhinglish) turns Roman-script Hinglish (how Indians actually type) into clean Devanagari for Indian TTS pipelines — deterministic, CPU-only, MIT-licensed.

Run the cells below (Runtime → Run all). No GPU, no API key needed.

```python
from openhinglish import normalize

result = normalize("bhai kal mera intv h paytm me")
print("display:", result.display)   # for humans
print("tts    :", result.tts)       # for a TTS engine
print("conf   :", round(result.confidence, 2))
```

```python
# Per-token detail + explainability trace
for tok in result.tokens:
    print(f"{tok.surface:8} {tok.category.name:12} display={tok.display_form:10} tts={tok.tts_form:10} conf={tok.confidence:.2f}")
    for step in tok.trace:
        print("        ", step)
```

```python
# Now try your own Hinglish:
from openhinglish import normalize, Config

text = "yaar mujhe 2 din me reply karo"  # <- change this
r = normalize(text)
print("display:", r.display)
print("tts    :", r.tts)

# Spell numbers as English words instead of Hindi:
r2 = normalize(text, config=Config(number_words_lang="english"))
print("tts (english numbers):", r2.tts)
```
